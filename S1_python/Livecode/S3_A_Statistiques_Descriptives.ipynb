## Visualisation univariée - 20 min

Installation de Seaborn si pas encore installé : pip install seaborn

```python
import seaborn as sns
import matplotlib as plt
import pandas as pd
```

```python
titanic = sns.load_dataset("titanic")
age = titanic.age.fillna(titanic.age.mean())
```

Variables qualitatives

On peut produire un tableau de fréquence pour une variable qualitative très facilement grâce à la commande :

pd.value_counts()

Afin de visualiser graphiquement le tableau de fréquence des différentes modalités d’une variable explicative qualitative, vous pouvez utiliser la commande suivante :

seaborn.countplot()

Variables quantitatives

Nous avons vu comment se passait la visualisation d’une distribution pour des variables catégoriques, mais il serait bon aussi de connaître la façon de visualiser la distribution de variables quantitatives.

Cette fois, nous pouvons faire ceci grâce à la fonction distplot()

###Distribution

```python
# Pour visualiser la distribution de l’âge dans le Titanic, on peut le faire de la façon suivante :
sns.distplot(age)
```

```python
# Dans notre graphique, nous avons obtenu, par défaut, environs 2 ans par barre dans le graphique. 
# Cependant, ceci n’est pas forcément la meilleure visualisation. On peut choisir le nombre de barres (bins en anglais) en le spécifiant dans les paramètres.

sns.distplot(age, bins = 10)
```

```python
# On peut aussi choisir de n’avoir qu’un histogramme et ne pas avoir le Kernel Density Estimation. De cette manière, on pourra voir exactement le nombre d’observations.

sns.distplot(age, kde=False, bins=10)
```

###Barplots

```python
town = titanic.embark_town.fillna(titanic.embark_town.mode())
```

```python
pd.value_counts(town)
```

```python
sns.countplot(town)
```

## Visualisation bivariée - 20 min

### Nuage des points

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(x="total_bill", y="tip", data = tips)
```

####Visualiser plus de deux dimensions sur un nuage de points

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", size="size" ,data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", size="size", style ="time", data=tips)
```

###Visualisation de relations continues

```python
df = pd.DataFrame(dict(time=np.arange(500), value=np.random.randn(500).cumsum()))
df.head()
```

```python
sns.relplot(x="time", y="value", kind="line", data=df)
```

###Voir une estimation

```python
irm = sns.load_dataset("fmri")
irm.head()
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=irm)
```

####Ajouter des dimensions

```python
sns.relplot(x="timepoint", y="signal", kind="line", hue="event", style="region", data=irm)
```

###Nuage de points par catégorie

```python
# Plus haut, nous avons utilisé la fonction relplot() pour toutes les relations continues. Pour les variables catégoriques, on utilise catplot().

import seaborn as sns
import matplotlib as plt
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.catplot(x="day", y="total_bill", data=tips)
```

```python
sns.catplot(x="total_bill", y="day", kind="swarm", data=tips)
```

####Ajouter des catégories

```python
sns.catplot(x="total_bill", y="day", hue="time", kind ="swarm", data=tips )
```

###Boxplot

```python
# La boite à moustache sera plus adapté en permettant de visualiser la distribution de chaque catégorie.

sns.catplot(x="day", y="total_bill", kind="box", data=tips)
```

####Ajouter des dimensions

```python
sns.catplot(x="day", y="total_bill", kind="box", hue="smoker", data=tips)
```

###Optimiser la visualisation de la distribution

```python
sns.catplot(x="day", y="total_bill", kind="boxen", hue="smoker", data=tips)
```

###Diagramme en barres / Bar plots

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.catplot(x="sex", y="survived", kind="bar", data=titanic)
```

####Ajouter une dimension avec hue

```python
sns.catplot(x="sex", y="survived", kind="bar", hue="class", data=titanic)
```
