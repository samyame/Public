# IF conditions<a id="if"></a>

General structure

<code>if (test is true) then:
    execute action 
elif (new condition to be tested) then:
    execute different action
elif (new condition to be tested) then:
    execute different other
else (all the remaining options) then:
    execute some new action</code>

## One level / one condition<a id="if_one_level_one"></a>

```python
question = input("What is the color of the sun?")
```

```python
question
```

```python
if question == 'yellow':
    print('congrats')
else:
    print('wrong')
```

```python
if question == 'yellow':
    print("Congrats")
else:
    print('This is wrong')
```

## One level / multiple condition<a id="if_one_level_two"></a>

- Stops as soon as the first condition is met
- So even if multiple conditions are good, only one bloc is executed

```python
a = 40

if a > 30:
    print('a is greater than 30')
elif a < 60 and a >= 0:
    print('a is lower than 60 and positive')
else:
    print('a is negative')
```

```python
type(a)
```

```python
isinstance(a, int)
```

## Multiple levels (Nested IFs)<a id="if_multiple_levels"></a>

```python
# The following statement fails because a is a string
# Write a nested structure to check that a is an integer, before running

a = '40'

if a > 30:
    print('a is greater than 30')
elif a < 60 and a >= 0:
    print('a is lower than 60 and positive')
else:
    print('a is negative')
```

```python
if isinstance(a, int):
# other option is to do: type(a) == 'int'
    if a > 30:
        print('a is greater than 30')
    elif a < 60:
        print('a is lower than 60')
    else:
        print('else')
else:
    print('Your input is not an integer')
```

# FOR loops<a id="for"></a>

General structure:

<code>for cursor in iterable:
    execute a given action
    cursor is automatically incremented</code>

Lists, strings, dictionnary, ranges are all iterables

### Iterating over range<a id="for_range"></a>

```python
# Note about the range range(i, j)
# Starts at i included. Stops at j - 1. Increment is 1

for i in range(0,3):
    print(i)
    print("Value of i: {}".format(i))
```

```python
range(0,3)
```

### Iterating over list<a id="for_list"></a>

```python
list = ['piano', 'flute', 'violin']
```

```python
len(list)
```

```python
list = ['piano', 'flute', 'violin']

for x in list:
    print(x)
```

### Iterating over string<a id="for_string"></a>

```python
word = 'Nicolas'
```

```python
len(word)
```

```python
word = 'Nicolas'

for index, x in enumerate(word):
    print(index, x)
```

# WHILE loops<a id="while"></a>

Execute a statement while a condition is true

```python
word = 'Nicolas'
length = len(word) 
length
```

```python
# Initialize i
i = 0

while i < length:
    print(word[i])
    print(i)
    i = 1 + i
    # equivalent to i = i + 1
```

Not a great example because we know what the truth is (the condition is well determined : length = 7) <br>
Other example : ask the same question until we have the right answer

```python
question = input("Yes or No?")

while question not in ("Yes", "No"):
    question = input("Yes or No?")
```

Or you write an "infinite" loop with an explicit if statement to break out

```python
while True:
    question = input("Yes or No?")
    if question in ("Yes", "No"):
        break
```
