```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
fro = pd.read_table('fromage.txt', sep = '\t',header=0,index_col=0)
```

```python
fro
```

```python
fro.describe()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
fro = pd.DataFrame(sc.fit_transform(fro),columns = fro.columns)
```

```python
fro
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(fro, 'ward')
dn = dendrogram(Z)
```

```python
from scipy.cluster.hierarchy import fcluster

P = fcluster(Z, t=7, criterion='distance')
```

```python
fro = fro.drop(index = [15,16])
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(fro, 'ward')
dn = dendrogram(Z)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
fropca = pd.DataFrame(pca.fit_transform(fro),columns = ['PCA 1', 'PCA 2'])
```

```python
fropca1 = pd.concat([fropca,pd.DataFrame(P)],axis = 1)
```

```python
fropca1.columns = ['PCA 1','PCA 2','hu']
```

```python
sns.scatterplot(x = fropca1.iloc[:,0], y = fropca1.iloc[:,1], hue = fropca1.iloc[:,2])
```
