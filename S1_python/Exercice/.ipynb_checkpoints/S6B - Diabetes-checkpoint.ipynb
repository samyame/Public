```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df[(df.Insulin <= (3*df.Insulin.std())+df.Insulin.mean()) | (df.Insulin >= (3*df.Insulin.std())-df.Insulin.mean())]
```

```python
sns.boxplot(y ='Insulin', data = df)
```

```python
for i in df.columns[:-1]:
    sns.pairplot(df[[i, "Outcome"]], hue="Outcome", size=5)
```

```python
df2 = df.copy()
```

```python
for i in df.columns:
    df2 = df2[(df2[i] <= (3*df2[i].std())+df2[i].mean())&(df2[i] >= 0)]
```

```python
df2 = df2.fillna(df2.mean())
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X = df2.drop(['Outcome'], axis = 1)

X1 = sc.fit_transform(X)
```

```python
df[['Outcome']]
```

```python
X2 =pd.DataFrame(X1)
```

```python
for i in X2.columns:
    sns.distplot(X2[i], hist = True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,df2[['Outcome']], test_size = 0.3)
```

```python
import sklearn.naive_bayes as nb

naivebayes = nb.GaussianNB()
naivebayes.fit(X_train , Y_train)
```

```python
naivebayes.score(X_train,Y_train)
```

```python
naivebayes.score(X_test,Y_test)
```

```python
for i in df2.columns:
    sns.factorplot(x = i, y = 'Outcome', data = df2, ci = None, join = False)
    plt.show()
```

```python
from sklearn.ensemble import BaggingClassifier

model = BaggingClassifier()
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```
