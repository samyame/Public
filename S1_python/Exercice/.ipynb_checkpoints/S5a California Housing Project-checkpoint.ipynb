```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import f_regression
from geopy.geocoders import Nominatim
```

```python
from sklearn import datasets
data = datasets.fetch_california_housing()
```

```python
data.feature_names
```

```python
data
```

```python
data1 = pd.DataFrame(data= np.c_[data['data'], data['target']],
                     columns= data['feature_names'] + ['target'])
data1 = data1.rename(columns={"target": "Price"})
```

```python
sns.relplot(x = 'MedInc', y= 'Price', data = data1)
```

```python
fig, ax = plt.subplots(figsize = (20,6))
sns.lineplot(x = 'HouseAge', y = 'Price', data = data1)
```

```python
fig, ax = plt.subplots(figsize = (20,6))
sns.regplot(x = 'AveRooms', y = 'Price', data = data1[data1.AveRooms<10])
```

```python
fig, ax = plt.subplots(figsize = (20,6))
sns.regplot(x = 'AveBedrms', y = 'Price', data = data1[data1.AveBedrms])
```

```python
fig, ax = plt.subplots(figsize = (20,6))
sns.regplot(x = 'Population', y = 'Price', data = data1[data1.Population < 15000])
```

```python
fig, ax = plt.subplots(figsize = (20,6))
sns.regplot(x = 'AveOccup', y = 'Price', data = data1[data1.AveOccup < 100])
```

```python
X = data1[['MedInc',
  'HouseAge',
  'AveRooms',
  'AveBedrms',
  'Population',
  'AveOccup',
  'Latitude',
  'Longitude']]
Y = data1['Price']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2)
```

```python
sc_x = StandardScaler()
X_train1 = sc_x.fit_transform(X_train)
X_test1 = sc_x.transform(X_test)
```

```python
reg = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None,
         normalize=False)
reg.fit(X_train1,Y_train)
```

```python
reg.score(X_train1,Y_train)
```

```python
reg.score(X_test1, Y_test)
```

```python
reg.coef_
```

```python
for i in zip(list(data1.columns),list(reg.coef_)):
        print(i)
```

```python
Coef = pd.DataFrame(data = {'Coefficient':reg.coef_}, index = np.array(data1.columns[:-1]))
Coef.sort_values(by = 'Coefficient', ascending = False)
Coef
```

```python
fx, fy = f_regression(X_train1, Y_train)
```

```python
freg = pd.DataFrame(data = {'f-score':fx,'p-value':fy}, index = np.array(data1.columns[:-1]))
freg.sort_values(by = 'f-score', ascending = False)
```

```python
sns.catplot(x = 'f-score', y = 'index' ,data= freg.reset_index(), kind = 'bar')
```

```python
data2 = data1.sample(1500).reset_index()
```

```python
data2
```

```python
def lng(X):
    geolocator = Nominatim(user_agent="TAR", timeout = 30)
    location = geolocator.reverse(str(data1.Latitude[X])+','+str(data1.Longitude[X]))
    try:
        return location.raw['address']['city']
    except:
        return 'unknown'
```

```python
data2['location'] = data2['index'].apply(lng)
```

```python
data2[~data2.location.isna()]
```

```python
data4 = data1.groupby(['location'])['Price'].mean().sort_values(ascending = False).reset_index()
data4
```

```python
data3 = data2.groupby(['location'])['Price'].mean().sort_values(ascending = False).head(5)
sns.barplot(x = 'location',y = 'Price', data = data3.reset_index())
```

```python
dummy = pd.get_dummies(data2.location)
```

```python
data2 = pd.concat([data2, dummy],axis = 1)
```

```python
data3 = data1[['Longitude', 'Latitude']]
data3['latlo'] = data3.reset_index()['index'].apply(lambda x : str(data1.Longitude[x]) +',' +str(data1.Latitude[x]))
data3 = data3.drop_duplicates('latlo')
data3['location'] = data3.reset_index()['index'].apply(lng)
```

```python
data3 = data3[~data3.location.isna()]
data1['latlo'] = data1.reset_index()['index'].apply(lambda x : str(data1.Longitude[x]) +',' +str(data1.Latitude[x]))
data1 = pd.merge(data1, data3[['latlo','location']], how = 'left', on = 'latlo')
```

```python
data1[data1.location.isna()]
```

```python
X = data2.drop(columns = ['Price', 'location', 'Latitude', 'Longitude','index'])
Y = data2['Price']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2)
```

```python
sc_x = StandardScaler()
X_train1 = sc_x.fit_transform(X_train)
X_test1 = sc_x.transform(X_test)
```

```python
reg = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None,
         normalize=False)
reg.fit(X_train1,Y_train)
```

```python
reg.score(X_train1, Y_train)
```

```python
reg.score(X_test1, Y_test)
```

```python
reg.coef_
```

```python
Cof = pd.DataFrame(index = np.array(X_train.columns), data = reg.coef_).sort_values(by = 0, ascending = False).reset_index()
Cof = Cof.rename(columns = {'index':'location', 0:'Coef'})
```

```python
Cof = pd.merge(Cof, data3[['Latitude','Longitude', 'location']], how = 'left', on = 'location')
Cof = pd.merge(Cof, data4, how = 'left', on = 'location')
Cof = Cof.drop_duplicates('location')
```

```python
data4 = pd.merge(data4, data1[['Latitude', 'Longitude', 'location']], how = 'left', on = 'location')
```

```python
BBox = ((-124.255,-115.049,      
         32.621, 41.968))
ruh_m = plt.imread('Cal.png')
fig, ax = plt.subplots(figsize = (8,7))
ax.scatter(Cof.Longitude, Cof.Latitude, zorder=1, alpha= 0.2, c='b', s=Cof['Coef']/10000000000)
ax.set_title('Plotting Spatial Data on Riyadh Map')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(ruh_m, zorder=0, extent = BBox, aspect= 'equal')
```

```python
BBox = ((-124.211,-120.157,      
         37.011, 39.300))
ruh_m = plt.imread('SF.png')
fig, ax = plt.subplots(figsize = (20,15))
for i in range(len(data4[(data4.Latitude > 37.011) & (data4.Latitude < 39.300)])):
    ax.scatter(data4.Longitude[(data4.Latitude > 37.011) & (data4.Latitude < 39.300)].array[i], data4.Latitude[(data4.Latitude > 37.011) & (data4.Latitude < 39.300)].array[i], zorder=1, alpha= 0.2,c = 'r', s=data4.Price[(data4.Latitude > 37.011) & (data4.Latitude < 39.300)].array[i]**2*100, label = data4.location[(data4.Latitude > 37.011) & (data4.Latitude < 39.300)].array[i])
ax.legend()
ax.set_title('House Prices SF')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(ruh_m, zorder=0, extent = BBox, aspect= 'equal')
```
