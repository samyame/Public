```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer_df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
```

```python
cancer_df
```

```python
Y = pd.DataFrame({cancer.target_names[0]:cancer.target})
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(cancer_df,Y,test_size = 0.25)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_trans = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso

reg = LinearRegression()
lass1 = Lasso(alpha = 1)
lass01 = Lasso(alpha = 0.01)
lass001 = Lasso(alpha = 0.001)
```

```python
reg1 = reg.fit(X_trans,Y_train)
lass1 = lass1.fit(X_trans,Y_train)
lass2 = lass01.fit(X_trans,Y_train)
lass3 = lass001.fit(X_trans,Y_train)
```

```python
import numpy as np
dfr = pd.DataFrame({'params':reg1.coef_[0], 'index' : range(1,len(cancer_df.columns)+1), 'model' :'reg'})
las1 = pd.DataFrame({'params':lass1.coef_,
                     'model' :'lasso 1' ,
                     'index': range(1,len(cancer_df.columns)+1)})
las2 = pd.DataFrame({'params':lass2.coef_,
                     'model' :'lasso 0.1' ,
                     'index': range(1,len(cancer_df.columns)+1)})
las3 = pd.DataFrame({'params':lass3.coef_,
                     'model' :'lasso 0.01' ,
                     'index': range(1,len(cancer_df.columns)+1)})
```

```python
df = pd.concat([np.transpose(dfr),np.transpose(las1),np.transpose(las2),np.transpose(las3)], axis = 1)
```

```python
df2 = pd.concat([dfr,las1,las2,las3])
df2
```

```python
import seaborn as sns
sns.pointplot(x = 'index', y = 'params',data = df2, hue = 'model', height = 8, aspect = 5)
```
