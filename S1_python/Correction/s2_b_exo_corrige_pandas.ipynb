```python
# Par compte d'order ID distincts
# Output est une série
chipotle.groupby('item_name')['order_id'].count().sort_values(ascending=False).head()
```

```python
chipotle.groupby('item_name')['order_id'].count().sort_values(ascending=False).reset_index().head()
```

```python
# Sur la quantity
chipotle.groupby('item_name')['quantity'].sum().sort_values(ascending=False).reset_index().head()
```

```python
# Operations multiples sur les colonnes
multiple_group_by = chipotle.groupby('item_name').agg({'order_id':'count',
                                  'quantity':['sum', 'max']})
```

```python
multiple_group_by.head()
```

```python
multiple_group_by.columns
```

```python
for x in multiple_group_by.columns:
    print(x)
```

```python
for x in multiple_group_by.columns:
    print(x[0] + '_' + x[1])
```

```python
multiple_group_by.columns = [x[0] + '_' + x[1] for x in multiple_group_by.columns]
```

```python
multiple_group_by.head()
```

```python
multiple_group_by = multiple_group_by.reset_index()
multiple_group_by.head()
```

```python
multiple_group_by.sort_values(by = ['quantity_sum', 'order_id_count'], ascending = False).head()
```
